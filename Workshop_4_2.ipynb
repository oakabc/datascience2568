```python
import matplotlib.pyplot as plt

# Sample data for plotting
x = [1, 2, 3, 4, 5]  # Common X-axis values

y1 = [1, 2, 3, 4, 5]  # Y-axis values for the first line
y2 = [2, 3, 4, 5, 6]  # Y-axis values for the second line
y3 = [3, 4, 5, 6, 7]  # Y-axis values for the third line
```

```python
# Plotting the lines

# Adding title and labels
plt.title('Multiple Line Chart Example')
plt.xlabel('X axis label')
plt.ylabel('Y axis label')


plt.plot(x, y1, label='Line 1')  # Plot the first line
plt.plot(x, y2, label='Line 2')  # Plot the second line
plt.plot(x, y3, label='Line 3')  # Plot the third line

# Showing legend
plt.legend()
# Displaying the plot
plt.show()
```
